### 1. LOAD AND ANALYZE DATA

```python
import pandas as pd

#Load data
df2 = pd.read_csv('train.csv')
df2.head()
```

```python

#datatypes
print (df2.dtypes)
```

###### WE SEE THAT THERE ARE NO STRING IN THE DATA, SO WE CAN PROCEED WITH TRAINING

```python
#total amount of rows and colunms

total_rows=len(df2.axes[0])

print("total rows", total_rows )


total_cols=len(df2.axes[1])

print("total columns", total_cols )
```

### 2. TRAINING DATA

```python

#features: all coluns except cover type
X = df2.drop(['Cover_Type'], axis = 1)

print(X.shape)

X.head()

#labels : = cover type

y = df2['Cover_Type']

print (y.shape)

#Asign
X_train,y_train= X,y
```

## 3 Comparing scikit-Learn ML algorithms

#### (i) K-Nearest Neighbor

```python
#import algorithm
from sklearn.neighbors import KNeighborsClassifier

 
#CHOOSE A Kneighbors CLASIFIER ALGORITHM

clasifier1 =  KNeighborsClassifier()

#print defualt algorithm training parameters
print(clasifier1.fit(X_train,y_train))

#test algorithm on train data
accuracy = clasifier1.score(X_train, y_train)
print("Original accuracy Kneighbors   on train data",accuracy )






```

#### (ii) LOGISTIC REGRESSION

```python
#import
from sklearn.linear_model import LogisticRegression


#CHOOSE A LOGISTIC REGRESSION ALGORITHM

clasifier2 = LogisticRegression(C=1000)

#label


#use algorithm to train
print(clasifier2.fit(X_train,y_train))

#test algorithm on train data
accuracy = clasifier2.score(X_train, y_train)
print("Original accuracy LOGISTIC REGRESSION  on train data",accuracy )

```

#### (iii). DECISION TREE CLASSIFIERS

```python
#import DecisionTreeClassifier ALGORITHM
from sklearn.tree import DecisionTreeClassifier

clasifier3 = DecisionTreeClassifier()


#use algorithm to train

print(clasifier3.fit(X_train,y_train))

#test algorithm on test and train data
accuracy = clasifier3.score(X_train, y_train)
print("Original accuracy  DecisionTreeClassifier   on train data",accuracy )
```

#### (iv) SUPPORT VECTOR MACHINE

```python
#CHOOSE A SVM ALGORITHM
from sklearn.svm import SVC

clasifier4 =SVC()


#use algorithm to train
print(clasifier4.fit(X_train,y_train))

#test algorithm on test and train data
accuracy = clasifier4.score(X_train, y_train)
print("Original accuracy  SVM  on train data",accuracy )
```

#### (v) LINEAR DISCRIMINANT ANALYSIS

```python

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

#import LinearDiscriminantAnalysis ALGORITHM

clasifier5 = LinearDiscriminantAnalysis()


#use algorithm to train

print(clasifier5.fit(X_train,y_train))

#test algorithm on test and train data
accuracy = clasifier5.score(X_train, y_train)
print("Original accuracy  LinearDiscriminantAnalysis  on train data",accuracy )
```

#### (vi) GAUSSIAN 

```python
from sklearn.naive_bayes import GaussianNB

#GaussianNB
clasifier6 = GaussianNB()


#use algorithm to train

print(clasifier6.fit(X_train,y_train))

#test algorithm on test and train data
accuracy = clasifier6.score(X_train, y_train)
print("Original accuracy GaussianNB  on train data",accuracy )
```

#### (vii). MULTIPLE LAYER PERCEPTRON (MLP) NEURAL NETWORK

```python
#Neural network MLP

from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

clasifier7 = MLPClassifier()

clasifier7 = MLPClassifier(random_state=1, max_iter=100).fit(X_train, y_train)
accuracy = clasifier7.score(X_train, y_train)
print("Original accuracy Neural network MLP on train data",accuracy )
```

## 4. COMPARING ALGORITHMS TOGETHER

```python
from sklearn import model_selection

array = df2.values

#COMPARING ALL MODELS TOGETHER
#All features except cover type
X = array[:,0:55]

#label: cover type
Y = array[:,55]


# prepare models
models = []
models.append(('Logisticregression', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
           metric_params=None, n_jobs=None, n_neighbors=5, p=2,
           weights='uniform')))
models.append(('DecisionTrees', DecisionTreeClassifier()))
models.append(('Gaussian', GaussianNB()))
#models.append(('SVM', SVC()))

# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:

    cv_results = model_selection.cross_val_score(model, X, Y, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f" % (name, cv_results.mean())
    print(msg)
```

## 5. (OPTIONAL) USING KERAS NEURAL NETWORK FOR TRAINING

```python

from keras.utils import to_categorical

ycat = to_categorical(y)

y_trains= ycat
print (y_trains.shape)
```

```python
#Using a keras Neural Network

import keras
from keras.models import Sequential
from keras.layers import Dense



# 3 layer Neural network
model = Sequential()

#1st layer
model.add(Dense(16, input_dim=55, activation='relu'))

#2nd Layer
model.add(Dense(12, activation='relu'))

#output layer
model.add(Dense(8, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X_train, y_trains, epochs=1000, batch_size=64)
```

```python

# evaluate the keras model on train data
scores = model.evaluate(X_train, y_trains, verbose=0)
print( scores[1])
```
